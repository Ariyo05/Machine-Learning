```python
import pandas as pd
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.model_selection import train_test_split
# from sklearn import metrics
# from sklearn.linear_model import LinearRegression
# from xgboost import XGBRegressor
```

```python
boston_house_price = sklearn.datasets.load_boston()
print(boston_house_price)
```

```python
boston_house_data = pd.DataFrame(boston_house_price.data, columns = boston_house_price.feature_names)
```

```python
boston_house_data.head()
```

```python
boston_house_data['Price'] = pd.DataFrame(boston_house_price.target)
```

```python
boston_house_data.head()
```

```python
boston_house_data.isnull().sum()
```

```python
boston_house_data.shape
```

```python
boston_house_data.describe()
```

```python
boston_correlation = boston_house_data.corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(boston_correlation, cbar= 2, fmt='.1f', cmap = 'Greens', annot =True)
```

```python
X = boston_house_data.iloc[:, :-1].values
X
```

```python
y = boston_house_data.iloc[:, -1:].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from xgboost import XGBRegressor
regression = XGBRegressor()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict([[0.00632,	18.0,	2.31,	0.0,	0.538,	6.575,	65.2,	4.0900,	1.0,	296.0,	15.3,	396.90,	4.98]])
y_pred
```

```python
reg_pred = regressor.predict(X_test)
reg_pred
```
